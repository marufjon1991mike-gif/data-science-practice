#### 19. Tasodifiy sonlardan iborat boʻlgan massiv yarating, soʻng ularni quyida berilgan 4 xil usul yordamida musbat sonlarning butun qismini ajratib oling.

`💡Ishora` % (Sonning kasr qismi - qoldiq), np.floor, astype, np.trunc

````python
rand_arr = np.random.randn(10)
pos_rand_arr = rand_arr[rand_arr > 0]
print(np.floor(pos_rand_arr))
print(pos_rand_arr.astype(int))
print(np.trunc(pos_rand_arr))
print(pos_rand_arr - pos_rand_arr%1)
````

#### 25. NumPy kutubxonasi yordamida kecha, bugun va ertangi kun sanalarini aniqlang.

#### 29. [1, 2, 3, 4, 5] vektor berilgan, har bir qiymat orasiga ketma-ket 3 ta nol raqamini yozuvchi dastur tuzing.
